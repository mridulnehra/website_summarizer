```python
import requests
from bs4 import BeautifulSoup
```

```python
class website:
    url:str
    title:str
    text:str
    def __init__(self,url):
        self.url = url
        response = requests.get(url)
        soup = BeautifulSoup(response.content,'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevent in soup.body(["script","style","img","input"]):
            irrelevent.decompose()
        self.text = soup.body.get_text(separator ="\n",strip =  True)
        
    
    
```

```python
# Let's try one out

ed = website("https://edwarddonner.com")
print(ed.title)
print(ed.text)
```

```python
system_prompt = "you are an assistant that analyze the contents of a website \
and provides a short summary,ignoring text that might be navigation related.\
rspond in markdown."
```

```python
def user_prompt_for(website):
     user_prompt= f" you are looking ata website titled {website.title} "
     user_prompt += "the content of this website is as follow;\
     please provide a short summary of thsi website in markdown.\
     if it includes news or announcements,then summarize these too.\n\n"
     user_prompt += website.text
     return user_prompt
    
```

```python
def message_for(website):
    return [
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt_for(website)}
    ]
```

```python
model = "llama3.2"
```

```python
def summarize(url):
    Website = website(url)
    messages = message_for(Website)
    response = ollama.chat(model=model, messages=messages)
    return response['message']['content']
```

```python
summarize("https://edwarddonner.com")
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python

display_summary("https://edwarddonner.com")
```
